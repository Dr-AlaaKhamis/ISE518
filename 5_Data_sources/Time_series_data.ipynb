# Time-Series Sensor Data 

This notebook loads and visualizes time-series data from an Excel file containing sensor measurements relevant to condition monitoring and predictive maintenance.

### Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Load the excel file

```python
# Install dependency if not already installed:
# !pip install openpyxl
file_path = "data/time_series/time_series_data.xlsx"
xls = pd.ExcelFile(file_path)

# Check sheet names
xls.sheet_names
```

### Load data from the first sheet

```python
df = xls.parse(xls.sheet_names[0])
df.head()
```

### Visualize the time-series data

```python
# Set the time column as index
df.set_index('Time', inplace=True)

# Plot time-series sensor data
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(df.index, df['Accel X'], label='Accel X')
plt.plot(df.index, df['Accel Y'], label='Accel Y')
plt.plot(df.index, df['Accel Z'], label='Accel Z')
plt.title('Acceleration Over Time')
plt.ylabel('Acceleration (g)')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(df.index, df['Ang Rate YZ'], label='Angular Rate YZ')
plt.plot(df.index, df['Ang Rate XZ'], label='Angular Rate XZ')
plt.title('Angular Rate Over Time')
plt.ylabel('Rate (deg/s)')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(df.index, df['Temperature '], label='Temperature')
plt.title('Temperature Over Time')
plt.ylabel('Temperature (°C)')
plt.xlabel('Time (s)')
plt.legend()

plt.tight_layout()
plt.show()
```
