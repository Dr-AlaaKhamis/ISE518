### Load dataset from url into a DataFrame

```python
import pandas as pd

# Use the raw GitHub URL instead of a local path
url = "https://raw.githubusercontent.com/Dr-AlaaKhamis/ISE518/main/2_Statistics/data/bottling_maintenance_events.csv"

df = pd.read_csv(url)
df.head()
```

#### Frequency of failure modes (categorical)

```python
counts_modes = df['failure_mode'].value_counts()
counts_modes
```

```python
plt.figure(figsize=(6,4))
counts_modes.sort_index().plot(kind='bar')
plt.title('Failure Modes Frequency')
plt.xlabel('Failure Mode')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

#### Histogram of time to failure (numerical)

Interpretation tip: right‑skewed shapes are common in reliability data because many items fail later, with a long tail.

```python
plt.figure(figsize=(6,4))
plt.hist(df['time_to_failure_days'], bins=20)
plt.title('Time to Failure (days)')
plt.xlabel('Days')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

### Change the number of bins and see how the shape looks

```python
plt.figure(figsize=(6,4))
plt.hist(df['time_to_failure_days'], bins=40, color='skyblue', edgecolor='black')
plt.title('Time to Failure (days) with 40 bins')
plt.xlabel('Days')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

### Filter to a single line (A/B/C) and compare histograms

```python
for line_id in ['A', 'B', 'C']:
    plt.figure(figsize=(6,4))
    subset = df[df['line'] == line_id]
    plt.hist(subset['time_to_failure_days'], bins=20, alpha=0.7, color='orange', edgecolor='black')
    plt.title(f'Time to Failure for Line {line_id}')
    plt.xlabel('Days')
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```
